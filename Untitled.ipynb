```python
import numpy as np
import matplotlib.pyplot as plt


num_cashiers = 3             
arrival_rate = 1 / 2          
service_rate = 1 / 7          
simulation_time = 500         


np.random.seed(50)           
arrival_times = []          
service_times = []           
current_time = 0
while current_time < simulation_time:
    inter_arrival_time = np.random.exponential(1 / arrival_rate)
    current_time += inter_arrival_time
    arrival_times.append(current_time)
    service_times.append(np.random.exponential(1 / service_rate))


cashiers = [0] * num_cashiers 
queue = []                   
wait_times = []                

for arrival, service in zip(arrival_times, service_times):
  
    for i in range(num_cashiers):
        if cashiers[i] <= arrival:
            cashiers[i] = 0

    if 0 in cashiers:
        cashier_index = cashiers.index(0)
        cashiers[cashier_index] = arrival + service
        wait_times.append(0)  # Без ожидания
    else:
       
        earliest_cashier_time = min(cashiers)
        queue_wait_time = earliest_cashier_time - arrival
        wait_times.append(queue_wait_time)
       
        cashier_index = cashiers.index(earliest_cashier_time)
        cashiers[cashier_index] = earliest_cashier_time + service


average_wait_time = np.mean(wait_times)
utilization = sum(cashiers) / (num_cashiers * simulation_time)

print(f"Среднее время ожидания в очереди: {average_wait_time:.2f} минут")
print(f"Загрузка кассиров: {utilization * 100:.2f}%")

# График
plt.hist(wait_times, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title("Распределение времени ожидания")
plt.xlabel("Время ожидания (минуты)")
plt.ylabel("Частота")
plt.show()
```
